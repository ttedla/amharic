# የቃላት አገባብ በአረፍተነገራት

በክፍል ፭ የቃላትን ስፍራ በመላው ሰነድ ውስጥ የሚለይ ፕሮግራም ጽፈናል። በዚህ ክፍል የምንጽፈው ፕሮግራም የቃላትን ስፈራ አረፍተነገርን እንደ ጠቋሚ በመጠቀም ማየት የሚያስችል ነው። በምሳሌ ለማስረዳት የሚከተለውን ጽሁፍ ከስበሀት ገብረ እግዚአብሔር አምስት ስድስት ሰባት እና ሌሎች ታሪኰች ከሚለው መጽሐፍ ገጽ 116 የተወሰደውን ጽሁፍ እንመልከት። ኰ እንዴት እንደተጻፈ አስተውሉ (ኮ) እንደዚህ የሚጻፍ ይመስለኝ ነበር፤ መጽሀፉ ላይ የተጻፈው ኰ ነው። 
*** 
*አምስት ስድስት ሰባት ፤ አምስት ልጅ ሁለት እናት አባት። ሰይጣን ከእባብ ተዋሀደ ሄዋንን አሳሳተ፤ አሳተ። ከእፀበለስ ተዋሀደ፤ ሄዋን **ሆድ** ገባ። ከሆድዋ ተዋሀደ ለልጆችዋ ተረፈ። ሰይጣን ሲኦል ወረደ ማለት ወደ **ሆድ** ወረደ ማለት ነው። ከዚያስ? ከዚያማ ጭለማ ውስጥ ቁጭ ብሎ ይገዛን ያስሰግደን ጀመር። አምስት ስድስት ሰባት! ቀኑን ሙሉ ስንሰራ ስንለፋ ስንዋሽ ስናስመስል - ለምን? ለእንጀራ። ለማን? ለሰይጣን **ሆድ**። **ሆድ** ይኽ **ሆድ** ይኽ ጐተራ፤ ይኽ ጭለማ የሰይጣን ቤት መኖሪያ። ሰው ከተወለደ እስኪሞት ተራራ ያህል እህል ይበላ። አይሞላ ታድያ ያ ሆዱ ያ ጉዱ። አይሞላም ሰይጣን ነው። **ሆድ**! **ሆድ**! **ሆድ**! ብሎ እየጮኸ ሆዱን በእጁ ደጋግሞ ሲፀፋው ተሰማቸው አባ ወልደማርያም።*   
***

## መግቢያ

ከላይ በተመለከትነው የጋሽ ስብሀት ቅንጭብ ጽሁፍ ውስጥ በአራት ነጥብ የሚያልቅ አስራሦስት አረፍተነገራትን ይዟል።  ይህ ተቀንጭቦ የተውሰደው ጽሁፍ *ሆድ* የሚለውን ጥሬ ቃል የያዙ ፭ አረፍተነገራት አሉ። እነሱም አረፍተነገር \[3, 5, 8, 9, 13\] ናቸው። ጥሬ ቃል ያልኩበት በጽሁፉ ውስጥ ሆድዋ፣ ሆዱ የሚሉ ያልተቆጠሩ ቃላት አሉ። መነሻ ግንዳቸው ከሆድ ጋር ተመሳሳይ የሆነ። ወደሌላ ርዕስ ውስጥ ላለመገባት ጥሬ ቃሉን ብቻ እንዳለ በመውሰድ እንዴት ትንተናዎችን መስራት እንደምንችል እንመልከት። 

ስለዚህም ሆድ የሚለው ቃል ጥቅም ላይ የዋለበትን አግባብ ወይም አውድ ማየት ብንፈልግ እነዚህን አረፈተነገራት (ማለትን አነገር 3፣ 5፣ 8፣ 9 እና 13ን) ለይተን መመልከት እንችላለን ማለት ነው። 

እስቲ ይሔንን ፕሮግራም ጽፈን እንመልከት። 

```python
# ቅንጭቡን ቅንጭብ በሚል ሕበረቁምፊ በሚይዝ ተለዋዋጭ እንወክለው። 
# በፓይተን \ ወደቀጣይ መስመር መተላለፉን አመላካች ነው። 
ቅንጭብ = "አምስት ስድስት ሰባት ፤ አምስት ልጅ ሁለት እናት አባት። \
        ሰይጣን ከእባብ ተዋሀደ ሄዋንን አሳሳተ፤ አሳተ። ከእፀበለስ ተዋሀደ፤ ሄዋን ሆድ ገባ።\
        ሆድዋ ተዋሀደ ለልጆችዋ ተረፈ። ሰይጣን ሲኦል ወረደ ማለት ወደ ሆድ ወረደ ማለት ነው።\
        ከዚያስ? ከዚያማ ጭለማ ውስጥ ቁጭ ብሎ ይገዛን ያስሰግደን ጀመር።\
        አምስት ስድስት ሰባት! ቀኑን ሙሉ ስንሰራ ስንለፋ ስንዋሽ ስናስመስል - ለምን? ለእንጀራ።\
        ለማን? ለሰይጣን ሆድ። ሆድ ይኽ ሆድ ይኽ ጐተራ፤ ይኽ ጭለማ የሰይጣን ቤት መኖሪያ።\
        ሰው ከተወለደ እስኪሞት ተራራ ያህል እህል ይበላ። አይሞላ ታድያ ያ ሆዱ ያ ጉዱ።\
        አይሞላም ሰይጣን ነው። \
        ሆድ! ሆድ! ሆድ! ብሎ እየጮኸ ሆዱን በእጁ ደጋግሞ ሲፀፋው ተሰማቸው አባ ወልደማርያም።"
ቅንጭብ
```

```python
# በመካከል የተፈጠረውን ካንድ እና ከዛ በላይ የሆኑ ተከታታይ ባዶ ቦታዎችን እንደሚከተለው መቀነስ እንችላለን 
# join() አገናኝ ተግባር ነው። ስለ ጆይን ተጨማር ማንበብ ትችላላችሁ።  
ቅንጭብ = " ".join(ቅንጭብ.split())
ቅንጭብ
```

```python
# በመቀጠል የማንፈልጋቸውን (ለትንተናችን ጠቃሚ ያልሆኑ ምልክቶችን) ሰንጠረዥ በመጠቀም እናስወግድ 
ሰንጠረዥ = str.maketrans("\t«»!፣፧፡፤?-", 10*(" "))
ቅንጭብ_የተጣራ = [ቅ.translate(ሰንጠረዥ) for ቅ in ቅንጭብ.splitlines() if ቅ]
# ውጤቱን እንመልከት 
ቅንጭብ_የተጣራ
```

```python
#ቅንጭብ_የተጣራን በአረፍተነገር እንከፋፍለው። ይህንን ለማድረግ አራት ነጥብን ለመክፈያነ እንጠቀማለን።  
ቅ_አነገር = [አነገር.split('።') for አነገር in ቅንጭብ_የተጣራ]
ቅ_አነገር # ቅ_አነገር የዝርዝር ዝርዝር ነው (ማለትም ሊስት ኦፍ ሊስት ነው። )
```

```python
#ቅ_አነገር የዝርዝር ዝርዝር ነው። 
# የመጀመሪያውን አረፍተነገር ለማግኘት ብንፈልቅ ሁለት ጠቋሚዎችን መጠቀም ይኖርብናል። 
# ቅ_አነገር[0][0]
ቅ_አነገር[0][0]
```

```python
# ከዝርዝር ዝርዝርነት ወደ ነጠላ ዝርዝር እንደሚከተለው እንቀይረው 
ቅ_አነገር_የተጣራ = [ዝ_አነገር for ዝዝ_አነገር in ቅ_አነገር for ዝ_አነገር in ዝዝ_አነገር]
ቅ_አነገር_የተጣራ
```

```python
# አሁን የመጀመሪያ አረፍተነገሩን ብንፈልግ 
ቅ_አነገር_የተጣራ[0]
```

```python
# አሁን የአረፍተነገር መዝገበ-ቃል መስራት እንችላለን። 
from collections import defaultdict
አነገር_መቃል = defaultdict(set)
```

```python
# በመቀጠል የአረፍተነገር መዝገበ-ቃሉን በ ቃላት እና በሚገኙበት አረፍተነገር መሙላት ነው። 
# ይህንን ለማድረግ ኢኑምሬትን እንጠቀማለን 
# አረፍተነገር እና ጠቋሚውን ለማግኘት  
for ጠቋሚ,አነገር in enumerate(ቅ_አነገር_የተጣራ): 
    #ቃላትን ለማግኘት አነገርን ባዶ ቦታ ሲገኝ መክፈል 
    for ቃል in አነገር.split():
        # አረፍተ ነገር ጠቋሚውን መዝገበቃሉ ውስጥ በቃሉ ቁልፍ እንደ እሴት መጨመር 
        አነገር_መቃል[ቃል].add(ጠቋሚ)
```

```python
# ሆድ ከላይ እንደተመለከትነው የሚገኘው በ ፭ አረፍተነገሮች ውስጥ ነው። 
# እነሱም አነገር 3፣ 5፣ 8፣ 9፣ 13
# የፓይተን ቁጥር የሚጀምረው ከ ዜሮ ስለሆነ፤ ማለትም የመጀመሪያውን አነገር 0 ብሎ ስለሚፈርጅ 
# 'ሆድ' የሚለው ቃል የሚገኝበት አነገር 2፣ 4፣ 7፣ 8፣ 12 ይሆናሉ ማለት ነው። 
# በፕሮግራም እንሞክረው። 
አነገር_መቃል['ሆድ']
```

```python
# አነገራቱን ማየት ብንፈልግ አነገራቱን የያዘው ዝርዝርን በመጠቀም ማየት ይቻላል። 
# አነገራቱን የያዘው ዝርዝር ቅ_አነገር_የተጣራ የሚለው ዝርዝር ነው። 
የሆድ_ጠቋሚ = [2, 4, 7, 8, 12]
for ጠቋሚ in የሆድ_ጠቋሚ:
    print(ቅ_አነገር_የተጣራ[ጠቋሚ], "።")
```

## ተፈላጊውን ፋይል ከፍቶ ማጥራት

የማይፈለጉ ምልክቶችን፣ ትርጉም የሌላቸው ቃላትን እና ስርዓተነጥቦችን ለትንተና የማያስፈልጉትን ከጽሁፉ ማስወገድን ይመለከታል። 

```python
#፩ ተፈላጊውን ፋይል ከፍቶ ማጥራት 

with open('./ውሂብ/አልወለድም_በአቤ_ጉበኛ.txt', 'r', encoding='utf-8-sig') as ዶሴ:
    ፋይል=ዶሴ.read().replace('\n', '')
    ሰንጠረዥ = str.maketrans("\t«»!፣፧፡", 7*(" "))
    ፋይል_የተጣራ = [ፋ.translate(ሰንጠረዥ) for ፋ in ፋይል.splitlines() if ፋ]
```

## የአረፍተነገር ዝርዝር ማዘጋጀት

ስራችንን ሰርተን ስንጨርስ የአረፍተነገራቱን ይዘት ማየት ብንፈልግ ከወዲሁ ማስቀመጣችን ይጠቅመናል። በዚህ ጽሁፍ አረፍተነገር የምንለው በ አራት ነጥብ የሚያበቃን ነው። አረፍተነገር በአማርኛ በ አራት ነጥብ ብቻ ነው የሚያበቃው? 

```python
# ፪፡ የአረፍተነገር ዝርዝር ማዘጋጀት 
ፋ_አነገር = [አነገር.split('።') for አነገር in ፋይል_የተጣራ]
ፋ_አነገር_የተጣራ = [ዝ_አነገር for ዝዝ_አነገር in ፋ_አነገር for ዝ_አነገር in ዝዝ_አነገር]
```

## ለቃላት የአረፍተነገር ማውጫ ማዘጋጀት

ይህ ማለት እያንዳንዱ ቃል የሚገኝበትን የአረፍተነገር ጠቋሚ ከቃሉ ጋር አጣምሮ ማስቀመጥ ማለት ነው። ይህንን ጠቋሚ በመጠቀምም በሰነዱ ውስጥ የሚገኙ ቃላቶች በየትኛው አረፍተነገር ውስጥ ጥቅም ላይ እንደዋሉ ለይተን ለማሳየት እንዲረዳን ነው። 

```python
# ፫፡ ነባሪ መዝገበቃል በመጠቀም የቃላት አረፍተነገር ማውጫ(ጠቋሚ) መስራት 
from collections import defaultdict
አነገር_መቃል_ጠቋሚ = defaultdict(set)
for ጠቋሚ,አነገር in enumerate(ፋ_አነገር_የተጣራ): 
    #ቃላትን ለማግኘት አነገርን ባዶ ቦታ ሲገኝ መክፈል 
    for ቃል in አነገር.split():
        # አረፍተ ነገር ጠቋሚውን መዝገበቃሉ ውስጥ በቃሉ ቁልፍ እንደ እሴት መጨመር 
        አነገር_መቃል_ጠቋሚ[ቃል].add(ጠቋሚ)
```

```python
# ፬፡ የቃሉን የአረፍተነገር ጠቋሚ በመጠቀም ቃሉን የያዙትን አረፍተነግራት ለይቶ ማሳየት 
# ለምሳሌ መጽሐፍ የሚለውን ቃል የያዙትን አነገራት እንለይ።
# መጽሐፍን የያዙ አነገራትን በዝርዝር ውስጥ ለማስቀመጥ ነው። 
መጽሐፍን_የያዙ_አነገሮች = []
መጽሐፍ_ጠቋሚዎች = አነገር_መቃል_ጠቋሚ['መጽሐፍ']
for መ_አነገሮች in መጽሐፍ_ጠቋሚዎች:
    መጽሐፍን_የያዙ_አነገሮች.append(ፋ_አነገር_የተጣራ[መ_አነገሮች]) 
```

```python
#መጽሐፍን_የያዙ_አነገሮች ብቻ በመጠቀም ውጤቱን ማየት እንችላለን። ነገር ግን ለእይታ እንዲያምር 
# እንደሚከተለው እናስተካክለው 
አነገራቱ =[(" ".join(t.split())+"።") for t in መጽሐፍን_የያዙ_አነገሮች]
for አነገር in አነገራቱ:
    print (አነገር, '\n')
```
